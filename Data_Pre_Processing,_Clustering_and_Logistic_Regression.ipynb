**ENCODING to convert categorical data into Numerical**

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'data' is your DataFrame
data_encoded = df.copy()

# Iterate over categorical columns and apply ordinal encoding
for col in data_encoded.columns:
    if data_encoded[col].dtype == 'object':
        oe = LabelEncoder()


        data_encoded[col] = oe.fit_transform(data_encoded[[col]])
```

```python
data_encoded.info()
```

**USING WRAPPER METHOD ( FORWARD SELECTION) to find Features which are optimal for BMI prediction**

```python
X = data_encoded.drop(['BMI'],axis=1)
y = data_encoded['BMI']
```

```python
import joblib
import sys
sys.modules['sklearn.externals.joblib'] = joblib

from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LinearRegression
# Sequential Forward Selection(sfs)
sfs = SFS(LinearRegression(),
          k_features=8,
          forward=True,
          floating=False,
          scoring = 'r2',
          cv = 0)
```

```python
sfs.fit(X, y)
sfs.k_feature_names_
```

***CLUSTERING ***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(data_encoded)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=1)
kmeans.fit(data_encoded)
x = data_encoded[('Weight')]
y = data_encoded[('BMI')]
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

**LOGISTIC REGRESSION**

```python
a = data_encoded.iloc[:, 10].values

b = data_encoded.iloc[:, 20].values
```

```python
X = np.array(a).reshape(-1, 1)
Y = np.array(b).reshape(-1, 1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(X_train)
xtest = sc_x.transform(X_test)

print (xtrain[0:10, :])
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain, y_train)
```

```python
y_pred = classifier.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print ("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```
